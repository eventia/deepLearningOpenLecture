```python
import numpy as np
```

```python
def AND(x1, x2):
    x = np.array([x1,x2])
    w = np.array([0.5, 0.5])
    b = -0.7
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    else :
        return 1
```

```python
def OR(x1,x2):
    x = np.array([x1,x2])
    w = np.array([0.5, 0.5])
    b = -0.2
    
    tmp = np.sum(w*x)+b
    if tmp <= 0:
        return 0
    else :
        return 1
```

```python
def NAND(x1,x2):
    x = np.array([x1,x2])
    w = np.array([-0.5, -0.5])
    b = 0.7
    
    tmp = np.sum(w*x)+b
    if tmp <= 0:
        return 0
    else :
        return 1
```

```python
def XOR(x1,x2):
    s1 = NAND(x1,x2)
    s2 = OR(x1,x2)
    y = AND(s1,s2)
    return y
```

```python
print(XOR(0,0), XOR(0,1), XOR(1,0), XOR(1,1))
```

```python
def XXOR(x1,x2):
    s2 = NAND(x1,x2)
    s4 = NAND(x1, s2)
    s5 = NAND(x2, s2)
    y = NAND(s4,s5)
    return y
```

```python
print(XXOR(0,0), XXOR(0,1), XXOR(1,0), XXOR(1,1))
```

```python
def AND(x1, x2):
    if (x1==1 and x2==1):
        return 1
    else :
        return 0
```

```python
print(AND(0,0), AND(0,1), AND(1,0), AND(1,1))
```
